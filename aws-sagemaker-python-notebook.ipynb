```python
# write function to be used to map tags within tag lists to moods
def tags_to_mood(tag):
    moods = [{'happy': ['karaoke', 'holiday', 'new year', 'classical']},
             {'heartbroken': ['heartbreak', 'heartbroken', 'tears']},
             {'angry': ['anger', 'angry', 'noise', 'horror']},
             {'in_love': ['love', 'kiss', 'romantic', 'ballad']},
             {'sad': ['sad', 'melancholy']},
             {'peaceful': ['chill', 'dream', 'peace', 'spiritual']},
             {'confident_sassy': ['trash', 'cybergrind', 'satire', 'cyberpunk', 'punk']},
             {'funky': ['funk']},
             {'dance': ['electro', 'electronic', 'break', 'hip-hop']}]
    for m_dict in moods:
        for mood, kw_tags in m_dict.items():
            for kw in kw_tags:
                if kw in tag:
                    return mood
            else:
                return ''
```

```python
genres_tags_df['mood'] = [tags_to_mood(tag) for tag in genres_tags_df['tags'].values]
```

```python
# for model validation, create train-test split
from sklearn.model_selection import train_test_split

X = moods_tags_df.title
y = moods_tags_df.mood

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=7)
```

```python
# inspect size and shape of train and test datasets
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# for asap baseline, will source sci-kit learn for cleaning and preprocessing of song titles
# note that ngram_range=(1,2) to get bi-grams
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(max_features=None, ngram_range=(1,2), stop_words='english')
vectorizer.fit(X_train)
print(vectorizer.vocabulary_)
```

```python
# complete vectorization of both X_train and X_test data
X_train_vectorized = vectorizer.transform(X_train)  # could convert to df but don't have to for ML modeling
X_test_vectorized = vectorizer.transform(X_test)  # same as with X_train_vectorized

X_train_vectorized.shape, X_test_vectorized.shape
```

```python
# use Logistic Regression modeling to check accuracy
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import random
```

```python
LR = LogisticRegression(random_state=7).fit(X_train_vectorized, y_train)

train_predictions = LR.predict(X_train_vectorized)
test_predictions = LR.predict(X_test_vectorized)
```

```python
# check both training and testing accuracy
print('Train Accuracy:', accuracy_score(y_train, train_predictions))
print('Test Accuracy:', accuracy_score(y_test, test_predictions))
```

```python
# use Word2Vec - credit to RA/LSDS
'''
Word2Vec is an increasingly popular word embedding technique. Like Bag-of-words it learns
a real-value vector representation for a predefined fixed-size vocabulary that is generated
from a corpus of text. However, in contrast to BoW, Word2Vec approaches are much more capable
of accounting for textual context, and are better at discovering words with similar meanings
or usages (semantic or syntactic similarity)
'''
!pip install -U gensim
import gensim
from gensim.models.word2vec import Word2Vec
```

```python
# fit track titles to Word2Vec
model = Word2Vec(titles, min_count=1, size=5)
```

```python
# Write a function that will receive as input a track title, split the title
# into a list of strings, send each string through the Word2Vec model for most-similar output word,
# find one or more tracks with that output_word, and get the mood of that/those track[s]

def recomm_by_title(track):
    # example track: 'Enter Sandman' by Metallica
    track_words = track.split(' ')
    for word in track_words:
        # print(word)
        try:
            output_list = model.wv.most_similar(word, topn=1)  # goal is to find single most-similar word
            output_word = output_list[0][0]  # output_list is a list with a single tuple inside it
            # print(output_word)
            # print(moods_tags_df.loc[moods_tags_df['title'].str.contains(word)]['title'])
            mood = moods_tags_df.loc[moods_tags_df['title'].str.contains(word)]['mood']
            print(mood)
            print()
            # here, need to ping DS django-REST-Heroku app with mood object above
        except Exception as e:
            print(e, 'occurred.', word, 'not found in title vocabulary.')
            continue


recomm_by_title('Enter Sandman')  # example of function call
```
