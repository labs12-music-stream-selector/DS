### Make MVP analysis dataframe and remove empty string and "[]" values. Then, engineer a "mood" feature and
### add this to end of the columns in the MVP analysis dataframe.

```python
# first, make a copy of the original dataframe created when data loaded in, for backup purposes
df = tracks_df.copy()
# make mvp dataframe
tracks_mvp_df = tracks_df[['track.5', 'track.19', 'track.7', 'track.8', 'track.9',
                           'track.18', 'artist.3', 'artist.12']]
# rename columns in mvp dataframe
tracks_mvp_df = tracks_mvp_df.rename(columns={'track.5': 'duration_in_seconds', 'track.19': 'title',
                                              'track.7': 'genre_top', 'track.8': 'genres',
                                              'track.9': 'genres_all', 'track.18': 'tags',
                                              'artist.3': 'bio', 'artist.12': 'artist_name'})
```

```python
# inspect mvp dataframe -- remove row indices 0 and 1
tracks_mvp_df = tracks_mvp_df.drop(tracks_mvp_df.index[[0,1]])
print(tracks_mvp_df.shape)
tracks_mvp_df.head(10)
```

```python
tracks_mvp_df.isna().sum()
```

```python
# make genres_tags_df, using specific genre_top values and tags values in features
genres_tags_df = tracks_mvp_df[(tracks_mvp_df['genre_top'].notna()) & (tracks_mvp_df['tags'] != '[]')]
```

```python
print(genres_tags_df.shape)  # tracks_df, the original, unchanged, df from the loaded data, is (106576, 53)
print('null count:', genres_tags_df.isna().sum())
genres_tags_df.head(10)
```

```python
# drop tracks with duplicate titles from genres_tags_df
genres_tags_df = genres_tags_df.drop_duplicates(subset=['title'])
```

```python
# add 'mood' feature to genres_tags_df, initializing all values as empty string
genres_tags_df['mood'] = ''
```

```python
# verify that mood feature now present
genres_tags_df.head()
```

```python
# write function to be used to map tags within tag lists to moods
def tags_to_mood(tag):
    moods = [{'happy': ['karaoke', 'holiday', 'new year', 'classical']},
             {'heartbroken': ['heartbreak', 'heartbroken', 'tears']},
             {'angry': ['anger', 'angry', 'noise', 'horror']},
             {'in_love': ['love', 'kiss', 'romantic', 'ballad']},
             {'sad': ['sad', 'melancholy']},
             {'peaceful': ['chill', 'dream', 'peace', 'spiritual']},
             {'confident_sassy': ['trash', 'cybergrind', 'satire', 'cyberpunk', 'punk']},
             {'funky': ['funk']},
             {'dance': ['electro', 'electronic', 'break', 'hip-hop']}]
    for m_dict in moods:
        for mood, kw_tags in m_dict.items():
            for kw in kw_tags:
                if kw in tag:
                    return mood
            else:
                return ''
```

```python
genres_tags_df['mood'] = [tags_to_mood(tag) for tag in genres_tags_df['tags'].values]
```
