```python
# Here I'm inferring a rating for each artist based on how many items each user listened to them using 'weight'

ratings = []

for row in users_and_songs['weight']:
    if row <= 100:
        ratings.append(1)
    elif row <= 200:
        ratings.append(2)
    elif row <= 300:
        ratings.append(3)
    elif row <= 400:
        ratings.append(4)
    elif row <= 500:
        ratings.append(5)
    elif row <= 600:
        ratings.append(6)
    elif row <= 800:
        ratings.append(7)
    elif row <= 1000:
        ratings.append(8)
    elif row <= 3000:
        ratings.append(9)
    else:
        ratings.append(10)
        
users_and_songs['songRatings'] = ratings
users_and_songs.head()
```

### Collaborative Filtering Using k-Nearest Neighbors
kNN method discovers clusters of similar users based on common ratings

```python
users_and_songs.head()
```

```python
song_ratingCount = (users_and_songs.
                   groupby(by=['song'])['songRatings'].
                   count().
                   reset_index().
                   rename(columns={'songRatings': 'totalRatingCount'})
                   [['song', 'totalRatingCount']]
                   )
song_ratingCount.head()
```

```python
song_ratingCount.head(25)
```

```python
rating_with_songRatingCount = song_ratingCount.merge(users_and_songs, left_on='song', right_on='song', how='left')
rating_with_songRatingCount.head()
```

```python
print(song_ratingCount['totalRatingCount'].describe())
```

The median artist has been rated only twice.  

```python
rating_with_songRatingCount.shape
```

### Implementing kNN

We convert our table to a 2D matrix, and fill the missing values with zeros(since we will calculate distances between rating vectors).  We then transform the values (ratings) of the matrix dataframe into a scipy sparse matrix for more efficient calculations.

### Finding the Nearest Neighbors
We use unsupervised algorithms with sklearn.neighbors.  The algorithm we use to compute the nearest neighbors is "brute", and we specify "metric=cosine" so that the algorithm will calculate the cosine similarity betwen rating vectors.  And then we fit the model.

```python
from scipy.sparse import csr_matrix
```

```python
rating_with_songRatingCount_pivot = rating_with_songRatingCount.pivot_table(index='song', columns='userID', values='songRatings').fillna(0)
rating_with_songRatingCount_pivot.head()
```

```python
rating_with_songRatingCount_matrix = csr_matrix(rating_with_songRatingCount_pivot.values)
```

```python
from sklearn.neighbors import NearestNeighbors

model_knn = NearestNeighbors(metric='cosine', algorithm='brute')
model_knn.fit(rating_with_songRatingCount_matrix)
```

### Test our model and make some recommendations:
In this step, the kNN algorithm measures distance to determine the "closeness" of instances.  It then classifies an instance by finding its nearest neighbors, and picks the most popular class among the neighbors.

```python
rating_with_songRatingCount_pivot.shape
```

```python
query_index = np.random.choice(rating_with_songRatingCount_pivot.shape[0])
query_index
```

```python
distances, indices = model_knn.kneighbors(rating_with_songRatingCount_pivot.iloc[query_index, :].values.reshape(1, -1), n_neighbors=6)

for i in range(0, len(distances.flatten())):
    if i == 0:
        print('Recommendations for {0}:\n'.format(rating_with_songRatingCount_pivot.index[query_index]))
    else:
        print('{0}: {1}, with distances of {2}:'.format(i, rating_with_songRatingCount_pivot.index[indices.flatten()[i]], distances.flatten()[i]))
```

### Collaborative Filtering Using Matrix Factorization

Matrix Factorization is simply a mathematical tool for playing around with matrices.  Matrix Factorization techniques are usally more effective because they allow users to discover the latent (hidden) features underlying the interactions between users and items (songs).

We use singular value decomposition (SVD) - one of the Matrix Factorization models for identifying latent factors.

Similar to kNN we convert our `rating_with_songRatingCount` table into a 2D matrix (called a utility matrix here) and fill the missing values with zeros.

```python
rating_with_songRatingCount_pivot_2 = rating_with_songRatingCount.pivot_table(index='userID', columns='song',values='songRatings').fillna(0)
rating_with_songRatingCount_pivot_2.head()
```

### We transpose this utility matrix, so that the name of songs becomes rows and userIDs become columns.  This is key.

```python
rating_with_songRatingCount_pivot_2.shape
```

```python
X = rating_with_songRatingCount_pivot_2.values.T
X.shape
```

```python
rating_with_songRatingCount_pivot_2.T.head()
```

```python
from sklearn.decomposition import TruncatedSVD

SVD = TruncatedSVD(n_components=12, random_state=42)
matrix = SVD.fit_transform(X)
matrix.shape
```

We calculate the Pearson's R correlation coefficient for every song pair in our final matrix.  To compare this with the results from kNN, we pick the same song: 'Desmeon - Undone' find the songs that have the highest correlation coefficients (between 0.9 and 1.0) with it.

```python
import warnings
warnings.filterwarnings("ignore", category=RuntimeWarning)
corr = np.corrcoef(matrix)
corr.shape
```

```python
rating_song_name = rating_with_songRatingCount_pivot_2.columns
rating_song_list = list(rating_song_name)
```

### So the results aren't as big as kNN, but still similar to it.  I also had to aim for a correlation between 1 and 0.7 to get any results most likely because of the small-size of the dataset.
